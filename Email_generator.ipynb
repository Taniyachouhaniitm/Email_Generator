```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    model="qwen/qwen3-32b",
    temperature=0
)


response = llm.invoke( """
Answer the question directly and concisely.
Do NOT show your reasoning.
Do NOT include thoughts, explanations, or analysis.
Only provide the final answer.

Question:
Who was the first person to go to the moon?
""")
print(response.content)
```

```python
from langchain_community.document_loaders import WebBaseLoader
loader=WebBaseLoader("https://careers.nike.com/software-engineer-iii-itc/job/R-76025")
page_data = loader.load().pop().page_content
print(page_data)
```

```python
from langchain_core.prompts import PromptTemplate

prompt_extract = PromptTemplate.from_template(
    """
    ### SCRAPED TEXT FROM WEBSITE:
    {page_data}
    
    ### INSTRUCTION:
    Answer the question directly and concisely.
    Do NOT show your reasoning.
    Do NOT include thoughts, explanations, or analysis.
    Do NOT Provide <think> </think>
    Only provide the final answer.
    The scraped text is from the career's page of a website.
    Your job is to extract the job post and return it in JSON format containing the following keys:
    'role', 'experience', 'skills', and 'description'.
    
    Only return the valid JSON.
    ### VALID JSON (NO PREAMBLE):
    """

)

chain_extract = prompt_extract | llm

res = chain_extract.invoke({
    "page_data": page_data
})

print(res.content)
```

```python
type(res.content)
```

```python
import re
from langchain_core.output_parsers import JsonOutputParser
text = res.content
clean_text = re.sub(r"<.*?>", "", text).strip()
json_parser = JsonOutputParser()
json_res = json_parser.parse(clean_text)
print(json_res)
```

```python
type(json_res)
```

```python
job = json_res
```

```python
import pandas as pd 
df = pd.read_csv("my_portfolio.csv")
df
```

```python
import uuid
import chromadb

client = chromadb.PersistentClient('vectorstore')
collection = client.get_or_create_collection(name="portfolio")

if not collection.count():
    for _, row in df.iterrows():
        collection.add(documents=row["Techstack"],
                        metadatas={"links": row["Links"]},
                        ids=[str(uuid.uuid4())])
        
```

```python
links= collection.query(query_texts=["Experience in python", "Experience in react"], n_results=1).get('metadatas')
links
```

```python
prompt_email = PromptTemplate.from_template(
"""
### JOB DESCRIPTION:
{job_description}

### INSTRUCTION:
You are Taniya, a Business Development Executive at XYZ Solutions, an AI & Software Consulting company.

Write a professional cold email strictly in the format shown below. Follow the structure, tone, and flow.

FORMAT TO FOLLOW:

Subject: <One concise professional subject related to the role>

Dear Hiring Manager,

<Opening paragraph: mention the role from job description and introduce XYZ Solutions>

<Second paragraph: describe technical capabilities aligned with the job>

<Portfolio paragraph: introduce 2–3 relevant examples using links from: {link_list}, in bullet points starting with * >

<Paragraph about company impact: scalability, optimization, cost reduction, efficiency>

<Closing paragraph: express interest in a discussion>

Best regards,
Taniya
Business Development Executive | XYZ Solutions

STRICT RULES:
- Use the name Taniya only.
- Use company name XYZ Solutions only.
- Do NOT use markdown, **, or special formatting.
- Do NOT add greetings like Hi.
- Do NOT add Best regards, name, or anything after the final line.
- Do NOT include contact details.
- Output ONLY the email.

### EMAIL:
"""
)


chain_email = prompt_email | llm

res = chain_email.invoke({
    "job_description": job,
    "link_list": links
})

import re
raw_text = res.content

clean_email = re.sub(r"<think>.*?</think>", "", raw_text, flags=re.DOTALL).strip()

print(clean_email)
```
