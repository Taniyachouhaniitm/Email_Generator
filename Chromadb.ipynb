```python
import chromadb
client = chromadb.Client()
collection = client.create_collection(name="my_collection")
```

```python
collection.add(
    documents=[
        "This document is about Newyork",
        "This document is about Delhi"
    ],
    ids=['id1','id2']
)
```

```python
all_documents=collection.get()
all_documents
```

```python
id1_document=collection.get(ids=['id1'])
id1_document
```

```python
results=collection.query(
    query_texts=['Query is about Chole Bhature'],
    n_results=1
)
results
```

```python
results=collection.query(
    query_texts=['Query is about Pizza'],
    n_results=1
)
results
```

```python
results=collection.query(
    query_texts=['Query is about Air Pollution'],
    n_results=1
)
results
```

```python
results=collection.query(
    query_texts=['Query is about India'],
    n_results=2
)
results
```

```python
collection.delete(ids=all_documents['ids'])
collection.get()
```

```python
collection.add(
    documents=[
        "This document is about Newyork",
        "This document is about Delhi"
    ],
    ids=['id1','id2'],
    metadatas=[
        {"url": "https://en.wikipedia.org/wiki/New_York_City"},
        {"url": "https://en.wikipedia.org/wiki/Delhi"}
    ]
)
```

```python
results=collection.query(
    query_texts=["Query is about Chole Bhature"],
    n_results=2
)
results
```
